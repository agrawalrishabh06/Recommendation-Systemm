```python
def fillna_by_distribution(df):
    for col in df.columns:
        missing_values = df[col].isna().sum()
        if missing_values:
            non_missing_values = df[col].dropna()
            sampled_values = non_missing_values.sample(missing_values, replace=True)
            sampled_values.index = df.loc[df[col].isna()].index
            df[col].fillna(sampled_values, inplace=True)

    return df

# Fill missing values and display the resulting DataFrame
filled_df = fillna_by_distribution(df)
```

```python
import pandas as pd
import numpy as np
import implicit
from scipy.sparse import csr_matrix
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.preprocessing import MaxAbsScaler
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.model_selection import GridSearchCV

# Create a custom estimator for the ALS model
class ALSWrapper(BaseEstimator, RegressorMixin):
    def __init__(self, factors=64, regularization=0.01):
        self.factors = factors
        self.regularization = regularization
        
    def fit(self, X, y=None):
        self.model = implicit.als.AlternatingLeastSquares(
            factors=self.factors, regularization=self.regularization
        )
        self.model.fit(X)
        return self
    
    def predict(self, X):
        user_factors = self.model.user_factors
        item_factors = self.model.item_factors
        return user_factors @ item_factors.T

    def get_params(self, deep=True):
        return {"factors": self.factors, "regularization": self.regularization}

# Assuming your user-item matrix is stored in the variable 'df'
data = np.array(df)

# Scale the data to improve model performance
scaler = MaxAbsScaler()
data_scaled = scaler.fit_transform(data)

# Create a sparse matrix representation of the data
sparse_data = csr_matrix(data_scaled)

# Initialize the ALS wrapper estimator
als_estimator = ALSWrapper()

# Define the hyperparameters to search over
param_grid = {
    "factors": [64, 32, 16],
    "regularization": [0.01, 0.1, 1.0, 10.0],
}

# Define the scoring metric as negative mean squared error
scoring = make_scorer(lambda y_true, y_pred: -mean_squared_error(y_true, y_pred))

# Perform a grid search over the hyperparameters
grid_search = GridSearchCV(als_estimator, param_grid, cv=5, n_jobs=-1, scoring=scoring)
grid_search.fit(sparse_data)

# Print the best hyperparameters and the corresponding MSE
print("Best Parameters: ", grid_search.best_params_)
print("Best Mean Squared Error: ", -grid_search.best_score_)
```

```python
## PMF 

import numpy as np

class PMF:
    def __init__(self, num_factors=10, num_epochs=100, learning_rate=0.01, regularization=0.1):
        self.num_factors = num_factors
        self.num_epochs = num_epochs
        self.learning_rate = learning_rate
        self.regularization = regularization

    def fit(self, train_data):
        num_users, num_items = train_data.shape

        self.user_factors = np.random.normal(scale=0.1, size=(num_users, self.num_factors))
        self.item_factors = np.random.normal(scale=0.1, size=(num_items, self.num_factors))

        for _ in range(self.num_epochs):
            for i in range(num_users):
                for j in range(num_items):
                    if train_data[i, j] > 0:
                        prediction = self.predict(i, j)
                        error = train_data[i, j] - prediction

                        self.user_factors[i] += self.learning_rate * (error * self.item_factors[j] - self.regularization * self.user_factors[i])
                        self.item_factors[j] += self.learning_rate * (error * self.user_factors[i] - self.regularization * self.item_factors[j])

    def predict(self, i, j):
        return np.dot(self.user_factors[i], self.item_factors[j])

    def predict_all(self):
        return np.dot(self.user_factors, self.item_factors.T)

```
